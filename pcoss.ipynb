```python
column_name_mapping = {
    'PCOS (Y/N)': 'PCOS',
    'Follicle No. (R)': 'Follicle_R',
    'Follicle No. (L)': 'Follicle_L',
    'Skin darkening (Y/N)': 'Skin_darkening',
    'hair growth(Y/N)': 'Hair_growth',
    'Weight gain(Y/N)': 'Weight_gain',
    'Cycle(R/I)': 'Cycle',
    'Fast food (Y/N)': 'Fast_food',
    'Pimples(Y/N)': 'Pimples',
    'AMH(ng/mL)': 'AMH',
    'Weight (Kg)': 'Weight',
    'BMI': 'BMI',
    'Hair loss(Y/N)': 'Hair_loss',
    'Waist(inch)': 'Waist',
    'Hip(inch)': 'Hip',
    'Endometrium (mm)': 'Endometrium'
}

# Rename the columns in the DataFrame
df_new= df_new.rename(columns=column_name_mapping)
```

```python
df_new.info()
```

```python
df_new.dropna(inplace=True)
```

```python
X=df_new.drop('PCOS',axis=1)
y=df_new['PCOS']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
predic=rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
print(classification_report(y_test,predic))
```

```python
accuracy_score(y_test,predic)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
pred1=lr.predict(X_test)
```

```python
accuracy_score(pred1,y_test)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Layer
from tensorflow.keras import optimizers
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import optimizers
```

```python
model = Sequential()
model.add(Dense(14,  activation='relu'))

model.add(Dense(7, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(X_train, y_train, epochs=200, batch_size=32, validation_data=(X_test, y_test))
```

```python
converter= tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model=converter.convert()
with open("model_pcos.tflite",'wb')as f:
    f.write(tflite_model)
```

```python
pred3 = (model.predict(X_test) > 0.5).astype("int32")
```

```python
print(classification_report(pred3,y_test))
```
